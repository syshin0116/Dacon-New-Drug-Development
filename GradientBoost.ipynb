```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./submissions/train_linear.csv')
df
```

```python
X_MLM = df.drop(columns=['MLM', 'id', 'SMILES'])
X_HLM = df.drop(columns=['HLM', 'id', 'SMILES'])
Y_MLM = df['MLM']
Y_MLM = df['HLM']
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import mutual_info_regression
from sklearn.ensemble import RandomForestClassifier
```

```python
bestfeatures = SelectKBest(score_func=mutual_info_regression, k=8)
fit = bestfeatures.fit(X_MLM, Y_MLM)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X_MLM.columns) 
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Features','Score']  
print(featureScores.nlargest(8,'Score'))
```
