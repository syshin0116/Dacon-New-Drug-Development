```python
import pandas as pd
from pycaret.regression import *

data = pd.read_csv('train.csv')
```

```python
data = data.drop(columns=['HLM'])
data
```

```python
from pycaret.regression import RegressionExperiment
s = RegressionExperiment()
s.setup(data=data,
                     target='MLM',
                     train_size= 0.8,
                     session_id=123,
                     fold=5)

setup(data=data,
                     target='MLM',
                     train_size= 0.8,
                     session_id=123,
                     fold=5)
```

```python
# functional API
best = compare_models()

# OOP API
best = s.compare_models()
```

```python
print(best)
```

```python
# functional API
evaluate_model(best)

# OOP API
s.evaluate_model(best)
```

```python
# functional API
evaluate_model(best)

# OOP API
s.evaluate_model(best)
```

```python
# functional API
plot_model(best, plot = 'residuals')

# OOP API
s.plot_model(best, plot = 'residuals')
```

```python
# functional API
plot_model(best, plot = 'feature')

# OOP API
s.plot_model(best, plot = 'feature')
```

```python
# functional API
predict_model(best)

# OOP API
s.predict_model(best)
```

```python
# functional API
predictions = predict_model(best, data=data)
predictions.head()

# OOP API
predictions = s.predict_model(best, data=data)
predictions.head()
```

```python
# functional API
save_model(best, 'my_best_MLM_pipeline')

# OOP API
s.save_model(best, 'my_best_MLM_pipeline')
```

```python
# load dataset
data = pd.read_csv('/content/test_linear.csv')

predictions_MLM = s.predict_model(best, data=data)
predictions_MLM.head()
```
